```python
import plotly.graph_objects as go
import pandas as pd
```

```python
df_bitcoin = pd.read_csv("Bitcoin Historical Data6.csv")
```

```python
df_bitcoin.info()
```

```python
df_bitcoin.head()
```

```python
# Convertim string a paràmetres numèrics
df_bitcoin['Price'] = df_bitcoin['Price'].str.replace(",", "").astype(float)
df_bitcoin['Open'] = df_bitcoin['Open'].str.replace(",", "").astype(float)
df_bitcoin['High'] = df_bitcoin['High'].str.replace(",", "").astype(float)
df_bitcoin['Low'] = df_bitcoin['Low'].str.replace(",", "").astype(float)
```

```python
# Convertim la data a tipus datetime
df_bitcoin['Date'] = pd.to_datetime(df_bitcoin['Date'], format='%m/%d/%Y')
```

```python
df_reverse_bitcoin_short = df_bitcoin.iloc[::-1]
```

```python
fig = go.Figure(
    data=go.Ohlc(
        x=df_reverse_bitcoin_short['Date'],
        open=df_reverse_bitcoin_short['Open'],
        high=df_reverse_bitcoin_short['High'],
        low=df_reverse_bitcoin_short['Low'],
        close=df_reverse_bitcoin_short['Price'] 
    ))
fig.update(
    layout_xaxis_rangeslider_visible=True
)
fig.update_layout(
    title="Evolució Preu Bitcoin (USD)",
    xaxis_title="Temps",
    yaxis_title="USD",
    width=1000,
    height=1000,
    paper_bgcolor="lightgray"
)
fig.show()
```
